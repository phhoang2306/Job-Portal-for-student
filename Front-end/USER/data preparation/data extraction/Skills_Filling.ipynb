```python
import pandas as pd
import numpy as np
import ast
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
```

```python
df = pd.read_csv("cleaned_data.csv")
```

```python
df.head(10)
```

```python
df['skills'] = df['skills'].apply(lambda x: ast.literal_eval(x))
df.skills[13920][1]
```

```python
vectorizer = TfidfVectorizer()
job_name_vectors = vectorizer.fit_transform(df['title'])
job_desc_vectors = vectorizer.fit_transform(df['Mô tả công việc'])
job_req_vectors = vectorizer.fit_transform(df['Yêu cầu ứng viên'])
```

```python
name_similarities = cosine_similarity(job_name_vectors)
desc_similarities = cosine_similarity(job_desc_vectors)
req_similarities = cosine_similarity(job_req_vectors)
```

```python
for i in range(len(df)):
    if not bool(df.loc[i, 'skills']):
        most_similar = np.argsort(similarity_scores[i])
        if len(most_similar) > 1:
            most_similar = most_similar[-2]
            while not bool(df.loc[most_similar, 'skills']):
                most_similar -= 1
                if most_similar < 0:
                    break
            if bool(df.loc[most_similar, 'skills']):
                df.loc[i, 'skills'] = df.loc[most_similar, 'skills']
            else:
                # If no similar row has non-missing skills, fill in with empty list
                df.loc[i, 'skills'] = []
        else:
            # If there is only one row, fill in with empty list
            df.loc[i, 'skills'] = []
```
